```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

import pandas as pd
```

```python
URL = 'https://markets.hankyung.com/consensus'
driver = webdriver.Chrome(executable_path='chromedriver')

driver.get(url=URL)
```

```python
nav = driver.find_element_by_xpath('//*[@id="container"]/div/div/article/section[2]/div[1]/a[2]')
nav.click()
```

```python
last_page = driver.find_element_by_xpath('//*[@id="container"]/div/div/article/section[3]/div[51]/nav/ul/li[8]/a').text

data = []
for i in range(int(last_page)):
    sleep(10) # 나중에 다른 방법으로 수정 필요
    for j in range(1, 51):
        root = '//*[@id="container"]/div/div/article/section[3]/div[{0}]/div[2]/a/div'.format(j)
        enterprise = driver.find_element_by_xpath(root + '/p[1]').text
        consensus = driver.find_element_by_xpath(root + '/ul/li/p[1]/span[2]').text
        brokerage = driver.find_element_by_xpath(root + '/ul/li/p[4]').text
        broker = driver.find_element_by_xpath(root + '/ul/li/p[3]').text
        title = driver.find_element_by_xpath(root + '/p[2]/a').text
        date = driver.find_element_by_xpath(root + '/ul/li/p[5]').text
        data.append([enterprise , consensus, brokerage[4:], broker[4:], title[5:], date[4:]])

    next_page = driver.find_element_by_xpath('//*[@id="container"]/div/div/article/section[3]/div[51]/nav/ul/li[9]/a')
    next_page.click()
```

```python
dataframe = pd.DataFrame(data) 
dataframe
```

```python
dataframe.to_csv('증권사리포트.csv', encoding='utf-8-sig') 
```
