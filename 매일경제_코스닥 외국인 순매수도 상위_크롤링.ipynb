```python
from selenium import webdriver
import pandas as pd
import re
from datetime import datetime
```

```python
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
page = 200
```

```python
result = pd.DataFrame(columns=['날짜', 'link'])
for i in range(page):
    url = f'http://find.mk.co.kr/new/search.php?pageNum={i+1}&cat=&cat1=&media_eco=&pageSize=10&sub=all&dispFlag=OFF&page=news&s_kwd=%5C%5C%5C%5C%BF%DC%B1%B9%B0%E8+%BC%F8%B8%C5%BC%F6%5C%5C%5C%5C&s_page=news&go_page=&ord=1&ord1=1&ord2=0&s_keyword=%5C%5C%BF%DC%B1%B9%B0%E8+%BC%F8%B8%C5%BC%F6%2C%B5%B5+%BB%F3%C0%A7%C1%BE%B8%F1%28%C4%DA%BD%BA%B4%DA%29+%B1%DD%BE%D7%B1%E2%C1%D8%5C%5C&s_i_keyword=%5C%5C%5C%5C%5C%5C%5C%5C%BF%DC%B1%B9%B0%E8+%BC%F8%B8%C5%BC%F6%5C%5C%5C%5C%5C%5C%5C%5C&s_author=&y1=1991&m1=01&d1=01&y2=2021&m2=07&d2=13&ord=1&area=tt'
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')

    link = soup.select('div.sub_list > span > a')
    cal = soup.select('div.sub_list > span.art_time')

    for i in range(len(cal)):
        cal[i] = cal[i].text[10:]
        cal[i] = cal[i][0:4]+cal[i][6:8]+ cal[i][10:12]
    
    for i in range(len(link)):
        link[i] = link[i]['href']
    
    link_list = pd.DataFrame({'날짜': cal, 'link':link})
    result = pd.concat([result,link_list])
```

```python
result_save = result
result = result.reset_index(drop=True)
result = result.loc[[i*4 for i in range(int(len(result)/4))]]
result["날짜"] = pd.to_datetime(result['날짜'])
```

```python
result["link"].tolist()
```

```python
buy_name_column= []
buy_value_column = []
sell_name_column= []
sell_value_column = []

for i in range(len(result)):    
    url = result["link"].tolist()[i]
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    p = re.compile('\d{6}')
    q = re.compile('\d{1,3}[.]\d{1,4}')
    rank = soup.select('td > .div_left')[0].text
    rank_name = p.findall(rank)
    rank_value = q.findall(rank)
    
    buy_name_column.append(rank_name)
    buy_value_column.append(rank_value)

    rank = soup.select('td > .div_right')[0].text
    rank_name = p.findall(rank)
    rank_value = q.findall(rank)
    
    sell_name_column.append(rank_name)
    sell_value_column.append(rank_value)
```

```python
result = result.reset_index(drop=True)
df1 = pd.DataFrame({'매수 종목':buy_name_column , '매수 금액': buy_value_column})
df2 = pd.DataFrame({'매수 종목':sell_name_column , '매수 금액':sell_value_column})
df_result = pd.concat([result,df1,df2], axis=1)
df_result_old = df_result
df_result = df_result.drop(['link'], axis=1)
```

```python
df_result.to_csv("df_Foreigner.csv", mode='w')
```
